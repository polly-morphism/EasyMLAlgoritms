```python
# Pre Processing
class NormStandart():
    def __init__(self, dataset, dsformat):
        self.dataset = dataset
        self.format = dsformat

    def normalize(self, dataset, rescale=1):
        max = dataset.max()
        min = dataset.min()
        for i in range(len(dataset.transpose()[0])):
            for j in range(dataset.shape[1]):
                dataset[i][j] = (dataset[i][j] - min)/(max-min)*rescale
        return dataset


    def mean(self, dataset):
        col_quant = len(dataset[0])
        dataset = dataset.transpose()
        means = np.zeros(col_quant)
        for i in range(col_quant):
            means[i] = np.sum(dataset[i])/float(len(dataset[i]))
        return means

    # calculate column standard deviations
    def stdev(self, dataset, means):
        col_quant = len(dataset[0])
        stdevs = np.zeros(col_quant)
        for i in range(col_quant):
            variance = [pow(row[i]-means[i], 2) for row in dataset]
            stdevs[i] = sum(variance)
        stdevs = [sqrt(x/(float(len(dataset)-1))) for x in stdevs]
        return np.array(stdevs)

    def standardize_dataset(self, dataset, means, stdevs):
        for row in dataset:
            for i in range(len(row)):
                row[i] = (row[i] - means[i]) / stdevs[i]

    def normalize_and_standartize(self):

        self.dataset = self.normalize(self.dataset)
        means = self.mean(self.dataset)
        stdevs = self.stdev(self.dataset, means)

        for row in self.dataset:
            for i in range(len(row)):
                row[i] = (row[i] - means[i]) / stdevs[i]
        if self.format == "numpy":
            return self.dataset
        elif self.format == "pandas":
            return pd.DataFrame(data=self.dataset)

ns = NormStandart(df.to_numpy(), "pandas")
df = ns.normalize(df.to_numpy(), 1000)
df = pd.DataFrame(data=df)
df.head()
```

```python
def predictQuality(x,theta):
    return np.dot(x,theta)

def calculateCost(x,theta,Y):
    prediction = predictQuality(x,theta)
    return ((prediction - Y)**2).mean()/2
```

```python
def abline(x,theta,Y, label_x, label_y, marginx, margin_y):
    """Plot a line from slope and intercept"""
    
    y_vals = predictQuality(x,theta)
    plt.xlim(marginx[0], marginx[1])
    plt.ylim(margin_y[0], margin_y[1])
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.gca().set_aspect(0.1, adjustable='datalim')
    plt.plot(x,Y,'.',x, y_vals, '-')
    plt.tight_layout()
    plt.show()
```

```python
def gradientDescentLinearRegression(alpha, iter, iter_size, index_x, index_y, label_x, label_y, marginx, margin_y):
    theta0 = []
    theta1 = []
    costs = []
    predictor = df[index_x]
    x = np.column_stack((np.ones(len(predictor)),predictor))
    Y = df[index_y]
    theta = np.zeros(2)
    for i in range(iter):
        pred = predictQuality(x,theta)
        t0 = theta[0] - alpha *(pred - Y).mean()
        t1 = theta[1] - alpha *((pred - Y)* x[:,1]).mean()
        
        theta = np.array([t0,t1])
        J = calculateCost(x,theta,Y)
        theta0.append(t0)
        theta1.append(t1)
        costs.append(J)
        if i%iter_size==0:
            print(f"Epoch: {i/iter_size}, Cost = {J}, theta = {theta}")
            abline(x,theta,Y, label_x, label_y, marginx, margin_y)
    print(f'theta0 = {theta0[-5:]}\ntheta1 = {theta1[-5:]}\nCosts = {costs[-5:]}')
```

```python
gradientDescentLinearRegression(0.0001,100, 25, 10, 11, "Alcocol Level", "Wine Quality", [15,35], [0, 25])
```
